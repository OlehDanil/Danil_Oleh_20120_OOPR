```

```
В каждом объекте этого класса должен быть публичный атрибут:

pole - двумерный кортеж, размером 3x3.

Каждый элемент кортежа pole является объектом класса Cell:
```
cell = Cell()
```
В объектах этого класса должно автоматически формироваться локальное свойство:

value - текущее значение в ячейке: 0 - клетка свободна; 1 - стоит крестик; 2 - стоит нолик.

Также с объектами класса Cell должна выполняться функция:

bool(cell) - возвращает True, если клетка свободна (value = 0) и False - в противном случае.

К каждой клетке игрового поля должен быть доступ через операторы:
```
res = game[i, j] # получение значения из клетки с индексами i, j
game[i, j] = value # запись нового значения в клетку с индексами i, j
```
Если индексы указаны неверно (не целые числа или числа, выходящие за диапазон [0; 2]), то следует генерировать исключение командой:
```
raise IndexError('некорректно указанные индексы')
```
Чтобы в программе не оперировать величинами: 0 - свободная клетка; 1 - крестики и 2 - нолики, в классе TicTacToe должны быть три публичных атрибута (атрибуты класса):
```
FREE_CELL = 0      # свободная клетка
HUMAN_X = 1        # крестик (игрок - человек)
COMPUTER_O = 2     # нолик (игрок - компьютер)
```
В самом классе TicTacToe должны быть объявлены следующие методы (как минимум):

computer_go() - реализация хода компьютера (ставит случайным образом нолик в свободную клетку).

````python
import random

class Cell:
    def __init__(self):
        self._value = 0

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, val):
        self._value = val

    def __bool__(self):
        return self._value == 0

class TicTacToe:
    FREE_CELL = 0
    HUMAN_X = 1
    COMPUTER_O = 2

    def __init__(self):
        self.pole = ((Cell(), Cell(), Cell()), (Cell(), Cell(), Cell()), (Cell(), Cell(), Cell()))

    def __getitem__(self, indexes):
        i, j = indexes
        if not isinstance(i, int) or not isinstance(j, int) or i < 0 or i > 2 or j < 0 or j > 2:
            raise IndexError('некорректно указанные индексы')
        return self.pole[i][j].value

    def __setitem__(self, indexes, value):
        i, j = indexes
        if not isinstance(i, int) or not isinstance(j, int) or i < 0 or i > 2 or j < 0 or j > 2:
            raise IndexError('некорректно указанные индексы')
        self.pole[i][j].value = value

    def init(self):
        for i in range(3):
            for j in range(3):
                self.pole[i][j].value = self.FREE_CELL

    def show(self):
        for i in range(3):
            for j in range(3):
                print(self.pole[i][j].value, end=' ')
            print()

    def human_go(self):
        while True:
            i, j = map(int, input("Введите координаты клетки для хода (i j): ").split())
            if self[i, j] == self.FREE_CELL:
                self[i, j] = self.HUMAN_X
                break
            else:
                print("Клетка занята, выберите другую.")

    def computer_go(self):
        empty_cells = [(i, j) for i in range(3) for j in range(3) if self[i, j] == self.FREE_CELL]
        i, j = random.choice(empty_cells)
        self[i, j] = self.COMPUTER_O

    @property
    def is_human_win(self):
        return self._check_win(self.HUMAN_X)

    @property
    def is_computer_win(self):
        return self._check_win(self.COMPUTER_O)

    @property
    def is_draw(self):
        return not any(cell.value == self.FREE_CELL for row in self.pole for cell in row)

    def _check_win(self, player):
        for i in range(3):
            if all(self[i, j] == player for j in range(3)):
                return True
            if all(self[j, i] == player for j in range(3)):
                return True
        if all(self[i, i] == player for i in range(3)):
            return True
        if all(self[i, 2-i] == player for i in range(3)):
            return True
        return False

    def __bool__(self):
        return not self.is_human_win and not self.is_computer_win and not self.is_draw
````
